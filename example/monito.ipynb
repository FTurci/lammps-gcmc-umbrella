```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt("n-betamu-3.092-T1.txt")
```

```python
plt.plot(data[:,1])
```

```python
from scipy.stats import sem
skip = 500
res = {'density' : data[skip:,1], 'pressure':data[skip:,2]}
nblocks = 5

for key,v in res.items():
    lenblock= len(v)//nblocks
    samples = [v[i*lenblock:(i+1)*lenblock].mean() for i in range(nblocks) ]
    print(key, np.mean(samples),sem(samples))
```
